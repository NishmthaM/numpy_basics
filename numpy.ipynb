```python
##rand() function

import numpy as np
rand= np.random.rand(4)
print(rand)

print()

rand1= np.random.rand(2,4)  ## 2 rows and 4 columns
print(rand1)

```

```python
import numpy as np
rand= np.random.rand(4)  ## same values but different numbers
print(rand)
```

```python
## randn() : Generates both p+ve and n-ve values

randn=np.random.randn(4)
print(randn)
```

```python
## ranf() : 1 is not included [0.1, 1.0)

var= np.random.ranf(5)
print(var)
```

```python
## randint() : give the values b/w the two min max values number
#randint(min,max,total_number)

var1= np.random.randint(2,6,3)
print(var1)
```

```python
## data types of numpy array

import numpy as np

var = np.array([1,2,3,4,5])
print("DataType :",var.dtype)
```

```python
## converting one data type into another data type

import numpy as np

x= np.array([1,2,3])
print("Data Type:",x.dtype)

x=np.array([1,2,3],dtype=np.int8)
print("Data type:",x.dtype)
```

```python
x1= np.array([1,2],dtype="f")
print(x1.dtype)

x2= np.array([1,2],dtype="c")
print(x2.dtype)
```
