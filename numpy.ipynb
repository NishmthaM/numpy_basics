```python
#Airthemetic operations in numpy in 1D array

#addition

import numpy as np

a=np.array([1,2,3,4])

varadd= a+3

print(varadd)
```

```python
a1=np.array([1,2,3,4])
a2=np.array([1,2,3,4])

varadd= a1+a2

print(varadd)
```

```python
#subtraction

b1=np.array([1,2,3,4])
b2=np.array([2,6,7,8])

varsub= b1-2

varsub2=b2-b1

print(varsub)
print(varsub2)

```

```python
#multiplication

c1=np.array([2,5,3])
c2=np.array([5,6,7])

varmulti1=c1*5
varmulti2=c2*5

varmulti3=c1*c2

print(varmulti1)
print(varmulti2)
print(varmulti3)
```

```python
#Division

d1=np.array([1,2,3,4])
d2=np.array([2,3,6,7])

vardiv1=d1/2
vardiv2=d2/2

vardiv3=d2/d1

print(vardiv1)
print(vardiv2)
print(vardiv3)
```

```python
#Modulus

e1=np.array([1,2.3,4])
e2=np.array([3,4,7,8,])

varmod1=e1%2
varmod2=e2%2

print(varmod1)
print(varmod2)
```

```python
##using built in functions

##Addition

a1=np.array([1,2,3,4])
a2=np.array([1,2,3,4])

varadd= np.add(a1,a2)

print(varadd)
```

```python
#subtraction

b1=np.array([2,2,3,4])
b2=np.array([2,6,7,8])

varsub=np.subtract(b2,b1)

print(varsub)

print()

#multiplication

c1=np.array([2,5,3])
c2=np.array([5,6,7])

varmulti=np.multiply(c1,c2)

print(varmulti)

print()

#Division

d1=np.array([1,2,3])
d2=np.array([2,3,6])

vardiv=np.divide(d1,d2)

print(vardiv)

print()

#Modulus

e1=np.array([1,2.3,4])

varmod=np.mod(e1,2)

print(varmod)

print()

##power

a1=np.array([12,3,4])

varpow=np.power(a1,3)

print(varpow)
```

```python
#Airthemetic operations in numpy in 2D array

#addition

import numpy as np

a=np.array([[1,2,3,4],[2,3,4,5]])
b=np.array([[2,5,4,8],[8,4,5,6]])

varadd= np.add(a,b)
print("2D addition")

print()

print(varadd)

print()

## 2D subtraction

c=np.array([[1,2,3,4],[2,3,4,5]])
d=np.array([[2,5,4,8],[8,4,5,6]])

varsub=np.subtract(d,c)

print("2D subtraction")
print()
print(varsub)

print()

##2D Multiplication

c1=np.array([[1,2,3,4],[2,3,4,5]])
d1=np.array([[2,5,4,8],[8,4,5,6]])

varmulti=np.multiply(c1,d1)

print("2D Multiplication")
print()
print(varmulti)

print()

##2D division

c2=np.array([[1,2,3,4],[2,3,4,5]])
d2=np.array([[2,5,4,8],[8,4,5,6]])

vardiv=np.divide(d2,c2)

print("2D division")
print()
print(vardiv)
print()

##2D Modulus

e=np.array([[2,3,4,5],[5,4,6,5]])

varmod=np.mod(e,5)

print("2D Modulus")
print()
print(varmod)
print()

```
