```python
#Airthemetic operations in numpy in 2D array

#addition

import numpy as np

a=np.array([[1,2,3,4],[2,3,4,5]])
b=np.array([[2,5,4,8],[8,4,5,6]])

varadd= np.add(a,b)
print("2D addition")

print()

print(varadd)

print()

## 2D subtraction

c=np.array([[1,2,3,4],[2,3,4,5]])
d=np.array([[2,5,4,8],[8,4,5,6]])

varsub=np.subtract(d,c)

print("2D subtraction")
print()
print(varsub)

print()

##2D Multiplication

c1=np.array([[1,2,3,4],[2,3,4,5]])
d1=np.array([[2,5,4,8],[8,4,5,6]])

varmulti=np.multiply(c1,d1)

print("2D Multiplication")
print()
print(varmulti)

print()

##2D division

c2=np.array([[1,2,3,4],[2,3,4,5]])
d2=np.array([[2,5,4,8],[8,4,5,6]])

vardiv=np.divide(d2,c2)

print("2D division")
print()
print(vardiv)
print()

##2D Modulus

e=np.array([[2,3,4,5],[5,4,6,5]])

varmod=np.mod(e,5)

print("2D Modulus")
print()
print(varmod)
print()

```

```python
#Airthemetic Functions
#1D array

import numpy as np

a=np.array([2,3,4,7,9,4,6,29,45,23,])

print("Minimum value =",np.min(a))
print("Maximum value =",np.max(a))
print("Position of min =",np.argmin(a))
print("Position of max =",np.argmax(a))
print("square root =",np.sqrt(a))

print()

##2D array

b=np.array([[2,6,4,],[9,5,4]])

print("Minimum value =",np.min(b,axis=0)) #min value in columns
print("Minimum value =",np.min(b,axis=1)) #min value in rows
print("Maximum value =",np.max(b,axis=0)) #max value in columns
print("Maximum value =",np.max(b,axis=1)) #max value in columns
print("Position of min =",np.argmin(a))
print("Position of max =",np.argmax(a))
print("Square root =",np.sqrt(b))
```
